```python
def roundOff(x, significant_digits):
    x = round(x, significant_digits - int(math.floor(math.log10(abs(x)))) - 1)
    return x
```

```python
def forwardEliminationWithPrecision(A, b, n, significant_digits, pivot=False, rounding=True):
    """
    Calculates the forward part of Gaussian elimination.
    """ 
    for currentPivotRow in range(n-1): # Iterate over rows - 1
        if pivot:
            # Check for Pivot elements
            maxindex = abs(A[currentPivotRow:, currentPivotRow]).argmax() + currentPivotRow
            
            if A[maxindex, currentPivotRow] == 0:
                raise ValueError("Matrix is singular.")  

            if maxindex != currentPivotRow: # Swap Rows
                A[[currentPivotRow, maxindex]] = A[[maxindex, currentPivotRow]]
                b[[currentPivotRow, maxindex]] = b[[maxindex, currentPivotRow]]
                
        else:
            if A[currentPivotRow, currentPivotRow] == 0: # Diagonal element
                raise ValueError("Pivot element is zero. Use Pivot.")
                
                
        for i in range(currentPivotRow + 1, n):
            factor = A[i, currentPivotRow] / A[currentPivotRow, currentPivotRow] # multiplier value
            if rounding:
                factor = np.around(factor, decimals = significant_digits)

            for j in range(currentPivotRow, n):
                A[i,j] = A[i,j] - factor * A[currentPivotRow,j]
                if rounding:
                    A[i,j] = np.around(A[i,j], decimals = significant_digits)

            b[i] = b[i] - factor * b[currentPivotRow]
            if rounding:
                b[i] = np.around(b[i], significant_digits)

        # print(f'\nIntermediate A at currentPivotRow {currentPivotRow} = \n%s and \n\nb = %s' % (A,b))
    return A, b

def backSubstitutionWithPrecision(a, b, n, significant_digits, rounding=True):
    """"
    Does back substitution, returns the Gauss result.
    """
    x = np.zeros((n,1))
    x[n-1] = b[n-1] / a[n-1, n-1]
    for currentRow in range(n-2, -1, -1):
        sums = b[currentRow]
        for j in range(currentRow + 1, n):
            sums = sums - a[currentRow,j] * x[j]
        x[currentRow] = sums / a[currentRow,currentRow]
        
    if rounding:
        for row in range(x.shape[0]):
            for col in range(x.shape[1]):
                x[row][col] = roundOff(x[row][col], significant_digits)
        
    return x

def gaussianEliminationWithPrecision(A, b, significant_digits, pivot=True, rounding=True):
    """
    This function performs Gauss elimination without pivoting.
    """
    n = A.shape[0] # Rows of Matrix A
    
    n = len(A)
    if b.size != n:
        raise ValueError("Invalid argument: incompatible sizes between" + "A & b.", b.size, n)

    A, b = forwardEliminationWithPrecision(A, b, n, significant_digits, pivot, rounding)
    return backSubstitutionWithPrecision(A, b, n, significant_digits, rounding)
```

```python
if __name__ == '__main__':
    
    m = 2
    n = 2
    A = matrixGeneration(m,n)
    B = matrixGeneration(m,1)
    
    equationFormulator(A, B)
    print()
    augmentedMatrix(A, B)
    print()
    
    significant_digits = [3, 4, 5, 6]
    
    pivots = [True, False]
    
    log = False
    for digits in significant_digits:
        for pivot in pivots:
            if log:
                print("\nOriginal A :\n", A)
                for row in range(A.shape[0]):
                    for col in range(A.shape[1]):
                        A[row][col] = roundOff(A[row][col], digits)
                print("\nAfter Rounding off :\n", A)

                print("\nOriginal B :\n", B)
                for row in range(B.shape[0]):
                    B[row][0] = roundOff(B[row][0], digits)
                print("\nAfter Rounding off :\n", B)

            print(f"Gaussian Elimination for significant_digits : {digits} with pivot equals {pivot}")
            x = gaussianEliminationWithPrecision(A, B, significant_digits=digits, pivot=pivot)
            print('\nGauss result is x = \n %s' % x)
            # print("\nGauss Elimination via library :\n", np.around(gaussEliminationLinAlg(A,B), digits))
            print()
```
